## Flow control

### loop

```python
for i in range(10):
    print(i)
```

```python
planetList = ["Earth", "Jupiter", "Mars"]
```

```python
for x in planetList:
    print(x, len(x))
```

```python
# use loop for a dictionary, then it will print the keys
for p in planet:
    print(p)
```

```python
for p in planet:
    print(f'The gravity of {p} is {planet[p]}.')
```

### condition

```python
p = "Mars"
```

```python
if planet[p] > 1:
    print(f'{p} has a lager gravity than the Earth.')
elif planet[p] < 1:
    print(f'{p} has a less gravity than the Earth.')
else:
    print(f'{p}has the same gravity as the Earth.')
```

```python
p = "Earth"
```

```python
if planet[p] > 1:
    print(f'{p} has a lager gravity than the Earth.')
elif planet[p] < 1:
    print(f'{p} has a less gravity than the Earth.')
else:
    print(f'{p} has the same gravity as the Earth.')
```

### Application

```python
#use loop and condition together!
for p in planet:
    if planet[p] > 1:
        print(f'{p} has a lager gravity than the Earth.')
    elif planet[p] < 1:
        print(f'{p} has a less gravity than the Earth.')
    else:
        print(f'{p} has the same gravity as the Earth.')
```
